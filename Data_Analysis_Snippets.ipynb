```
#8:09pm 08-01-2021
# ML
 
import pandas as pd
 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
 
 
data_frame = pd.read_csv('/content/music.csv') #data set courtesy of programming with Mosh
df = data_frame
music_data = df
md = music_data
 
#Attributes and Method
#df.shape
#df.describe()
#df.values
 
 
X = md.drop(columns=['genre'])
y = md['genre']

# Machine Learning
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.6) 
 
#X_train = X
#y_train = y
#X_test = specific input data in X-Dataframe format for prediction
#y_test = specific output prediction in y-Dataframe format
 
#test_size affects accuracy_score
 
 
 
 
 
 
model = DecisionTreeClassifier()

model.fit(X_train,y_train)
predictions = model.predict(X_test)

#3.0 Demonstrate
#print(model)
#print(f'\n{list(y_test)}\n{predictions}')
score = accuracy_score(y_test, predictions)

 
# Result
print(str(round(score*100,1))+' %', 'accuracy')
```

```
# Python program to generate WordCloud 
  
# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
  
# Reads 'Youtube04-Eminem.csv' file  
df = pd.read_csv(r"/content/clean_work.csv", encoding ="latin-1") 
  
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for key,val in df.iterrows(): 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    co = 0
    for i in range(len(tokens)):
        co+=1
        if co>=5:
          del tokens[i:]
          break
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 117), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
 
wordcloud.to_file('dental.png') 
plt.show()
```

```
#YouTube automation Script
import urllib, json, time
from selenium import webdriver #pip install selenium

def look_for_new_video():
	api_key = input("api_key?") #Add api_key, check https://console.developers.google.com
	channel_id = input("Channel_ID") #Add Channel ID
	
	base_video_url = 'https://www.youtube.com/watch?v='
	base_search_url = 'https://www.googleapis.com/youtube/v3/search?'
	
	url = base_search_url + f'key = {api_key}&channelID={channel_id}&part=snippet,id&order=date&maxResults=1'
	
	data = urllib.request.urlopen(url)
	resp = json.load(data)
	
	vidID = resp['items'][0]['id']['videoId']

	with open('videoid.json', 'r') as json_file:
		data = json.load(json_file)
		if data['videoId'] != vidID:
			driver = webdriver.Firefox()
			driver.get(base_video_url + vidID)
			video_exists = True
	
	if video_exists:
		with open('videoid.json', 'w') as json_file:
			data = {'videoId' : vidID}
			json.dump(data, json_file)

try:
	while True:
		look_for_new_video()
		time.sleep(10)
except KeyboardInterrupt:
	print('Alright, I\'m Stopping...')
```
